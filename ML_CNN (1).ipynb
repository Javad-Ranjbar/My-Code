```python
# import libraries
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from skimage import color
from skimage import io
```

```python
image = io.imread('image.jpg', as_gray=True) #importing image
plt.imshow(image, cmap = "gray") # converting to gray scale
plt.axis('off')
plt.show()
```

```python
from scipy import signal
filter1 = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]]) # filter1
CNN1 = signal.convolve2d(image,filter1) # assign filter 1 to grayscale picture
plt.imshow(CNN1, cmap = "gray")
plt.axis('off')
plt.show()
```

```python
filter2 = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]) #filter2
CNN2 = signal.convolve2d(image,filter2) # assign filter 2 to grayscale picture
plt.imshow(CNN2, cmap = "gray")
plt.axis('off')
plt.show()
```
